```python
from matplotlib import pyplot as plt
from ts_fun import corrcof
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
dfraw = pd.read_csv("./data/Inflation_data.csv") 
dfraw.head()

years = list(dfraw.columns[2:])
countries = list(dfraw["Country Name"])
X = []
cols = [] 
data_set = []
for idx,c in enumerate(countries):
    cols.append(c)
    data = []
    for y in years:
        data.append(dfraw[y][idx])
    data_set.append(data)



ds = pd.DataFrame(np.array(data_set).transpose(),columns=cols,index=years)
ds.index.name="Years"
print(ds.shape)
corrdf = pd.DataFrame(columns=countries,index=countries)
done = set()
for c1 in countries:
    for c2 in countries:
        if c1!=c2 and (c2,c1) not in done and (c1,c2) not in done:
            result =corrcof(ds[c2],ds[c1])
            corrdf[c2][c1] = result

            if result > 0.9 or result<-0.9:
                print(c2," <-> ",c1,"  ",result)
            if result < -0.8:
                print("Opposite",c1,c2)
            
            done.add((c1,c2))

corrdf.head()

```
